```python

def frac_bwn(T=2, dt=0.01, sigma=0.05, S0=0, mu=0.1, plot=False):
    N = round(T/dt)
    t = np.linspace(0, T, N)
    W = np.random.standard_normal(size = N) 
    W = np.cumsum(W)*np.sqrt(dt) ### standard brownian motion ###
    X = (mu-0.5*sigma**2)*t + sigma*W 
    S = S0*np.exp(X) ### geometric brownian motion ###
    
    if plot:
        plt.plot(t, S)
        plt.show()
    
    return t, S

fbm_t, fbm_s = frac_bwn(T=1, dt=0.001, plot=True)
print len(fbm_s)
```

```python
from scipy.stats import norm

# Process parameters
delta = 0.25
dt = 0.1

# Initial condition.
x = 0.0

# Number of iterations to compute.
n = 20

# Iterate to compute the steps of the Brownian motion.


def brown_m(n):
    # Process parameters
    delta = 0.25
    dt = 0.1

    # Initial condition.
    x = 0.0
    
    for k in range(n):
        x = x + norm.rvs(scale=delta**2*dt)
        yield x
bwn_m = np.array([i for i in brown_m(1000)])
plot(bwn_m)
    
```
